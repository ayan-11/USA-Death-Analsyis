```python
# Bar plot for 'Exceeds Threshold'
plt.figure(figsize=(6, 4))
pd.Series(data['Exceeds Threshold']).value_counts().plot(kind='bar')
plt.title('Count of Exceeds Threshold')
plt.xlabel('Exceeds Threshold')
plt.ylabel('Count')
plt.show()

```

```python
# Scatter plot for 'Upper Bound Threshold' vs 'Excess Estimate'
plt.figure(figsize=(8, 6))
plt.scatter(data['Upper Bound Threshold'], data['Excess Estimate'])
plt.title('Scatter plot: Upper Bound Threshold vs Excess Estimate')
plt.xlabel('Upper Bound Threshold')
plt.ylabel('Excess Estimate')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
plt.bar(data['State'], data['Observed Number'])
plt.title('Observed Number by State')
plt.xlabel('State')
plt.ylabel('Observed Number')
plt.xticks(rotation=45, ha='right') 
plt.show()
```
